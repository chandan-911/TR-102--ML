```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([2,3,4,6,7,9,7,5,1])
y=np.array([2,13,21,31,42,50,60,70,80])
```

```python
plt.plot(x,y)
```

```python
# marker 
plt.plot(x,y,marker ="*")
```

```python
# ms= mqarker size
plt.plot(x,y,marker ="*",ms=20)
```

```python
# marker color
# mec => marker edge color
# mfc=> marker face color

plt.plot(x,y,marker ="*",ms=20,mec ="red")
```

```python
plt.plot(x,y,marker ="*",ms=20,mfc="green")
```

```python
# 1,2,*,.,>,<,x,o
plt.plot(x,y,marker ="*",ms=20,mfc="green")
```

```python
# line ---> color 
plt.plot(x,y,marker ="*",ms=20,mfc="green",color="green")
```

```python
# thickness of line
plt.plot(x,y,marker ="*",ms=20,mfc="green",color="green",linewidth  =9)
```

```python
# line styles
# thickness of line
plt.plot(x,y,marker ="*",ms=20,mfc="green",color="green",linewidth  =9,linestyle ="dashed")
```

```python
# line styles
# thickness of line
plt.plot(x,y,marker ="*",ms=20,mfc="green",color="green",linestyle ="dotted")
```

```python
# line styles
# thickness of line
plt.plot(x,y,marker ="*",ms=20,mfc="green",color="green",linestyle ="dashdot")
```

```python
plt.plot(x,y,marker ="x",ms=15,mfc="green",color="green",linestyle ="dashdot")
```

```python
months=np.array([1,2,3,4,5,6,7])
profit=np.array([20,10,15,67,34,23,20])
```

```python
plt.plot(months,profit,marker=".",ms=6)
plt.xlabel("All Months")
plt.ylabel("My profits per month")
plt.title("my this year revenue",color= "blue")
```

```python
# apply grid
plt.plot(months,profit,marker=".",ms=6)
plt.xlabel("All Months")
plt.ylabel("My profits per month")
plt.title("my this year revenue",color= "blue")
plt.grid(color="yellow")
```

```python
# apply grid
plt.plot(months,profit,marker=".",ms=6)
plt.xlabel("All Months")
plt.ylabel("My profits per month")
plt.title("my this year revenue",color= "blue")
plt.grid(color="green",linestyle = "-.")
```
