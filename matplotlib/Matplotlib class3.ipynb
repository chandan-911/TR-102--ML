```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# bar chart
```

```python
months=np.array(["Jan","Feb","Mar","Apr","May"])
profit=np.array([10,6,7,9,6])
```

```python
# plt.bar(x,y)
plt.bar(months,profit)
```

```python
plt.bar(months,profit)
plt.xlabel("Months for profit")
plt.ylabel("profit per month")
plt.title("My revenue")
plt.grid()
```

```python
# change color of bars
colors=["red","blue","green","yellow","pink"]
plt.bar(months,profit,color=colors)
```

```python
# change color of bars
colors=["red","blue","green","yellow","pink"]
# x=[0.2,0.4,0.8,0.5,0.7]
plt.bar(months,profit,color=colors,width =x)
plt.xlabel("Months for profit")
plt.ylabel("profit per month")
plt.title("My revenue")
```

```python
# change color of bars
colors=["red","blue","green","yellow","pink"]
x=[0.2,0.4,0.8,0.5,0.7]
plt.barh(months,profit,color=colors,height=x)
plt.xlabel("Months for profit")
plt.ylabel("profit per month")
plt.title("My revenue")
```

```python
# change color of bars
colors=["red","blue","green","yellow","pink"]
x=[1.0,0.6,0.7,0.9,0.6]
plt.bar(months,profit,color=colors,width=x,label="Profit bars")

plt.plot(months,profit,color="brown",linestyle="dashdot",label ="profit trends")
plt.legend(loc="best")
# supported values are 'best', 'upper right', 'upper left', 'lower left',
# 'lower right', 'right', 'center left', 'center right', 'lower center',
# 'upper center', 'center'

# write labels for graph 
plt.xlabel("Months for profit")
plt.ylabel("profit per month")
plt.title("My revenue")

# write on bars 
for i in range(len(months)):
    plt.annotate(xy=(months[i],profit[i]),text=f"{profit[i]} Lakh")
```
